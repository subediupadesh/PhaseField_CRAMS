```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import matplotlib.patches as patches
from matplotlib.patches import Polygon 
from scipy.io import loadmat
from matplotlib.colors import ListedColormap
from matplotlib.lines import Line2D
from matplotlib.text import Text
```

## High Power Figure For Manuscript

```python
power_type = 'high_02' #high_power

path = os.path.abspath(f'../../../0_Moose_Files/{power_type}/exodus/')

model_path = path+'/FeNi.csv'
```

```python
df = pd.read_csv(model_path,  encoding='utf-8')
df['area_total'] =  df['area_h1']+df['area_h2']+df['area_h3']+df['area_h4']+df['area_h5']
df.head()
```

```python
df[292:293]
```

```python
df[1216:1217]
```

#### All area 

```python
t_step = len(df['time'])

fig, ax1 = plt.subplots(1,1, figsize=(10,8), frameon=False)
ax1.set_facecolor((0, 0, 0, 0))

TOT_area = ax1.plot(df['time'][:t_step], df['area_total'][:t_step], color='m', label=r'Total', linestyle='--', linewidth=7)
LIQ_area = ax1.plot(df['time'][:t_step], df['area_h1'][:t_step], color='red', label=r'LIQ ', linestyle='-', linewidth=7)
FCC_area = ax1.plot(df['time'][:t_step], df['area_h2'][:t_step]+df['area_h4'][:t_step], color='blueviolet', label=r'$\mathbf{\gamma}$', linestyle='-', linewidth=7)
BCC_area = ax1.plot(df['time'][:t_step], df['area_h3'][:t_step]+df['area_h5'][:t_step], color='pink', label=r'$\mathbf{\alpha}$', linestyle='-', linewidth=7)

ax1.legend(loc=(0.05, 0.7), ncol=2,  columnspacing=1.0, prop={'weight':'bold', 'size':'25',},  handletextpad=0.5, handlelength=2.0,   frameon=True)#

ax1.set_xlabel(r"t [$\mathbf{\mu s}$] $\mathbf{\longrightarrow}$", weight = 'bold', size=45, color='k')
ax1.set_ylabel(r"Area [$\mathbf{\mu m^2}$] $\mathbf{\longrightarrow}$", weight = 'bold', size=45, color='k')
ax1.get_yaxis().get_offset_text().set_position((-0.01, 0))

ax1.set_yticks([0, 1.0e4, 2.0e4, 3.0e4, 4.0e4], labels=[ '0', '1', '2', '3', '4'], size=35, color='k', rotation=0,  weight='bold')
ax1.set_title(r'$\mathbf{\times}$10$\mathbf{^{4}}$', pad=0, loc='left', color='black', fontsize=35, weight='bold')
ax1.set_xticks([0, 350, 700, 1050], labels=['0', '350', '750', '1050', ], size=30, color='k', weight='bold')

ax1.set_ylim(-0.4e4, 4.8e4)
ax1.set_xlim(-50, 1200)
ax1.tick_params(axis='y', direction='inout', which='both', length=15, width=5)
ax1.tick_params(axis='x', direction='inout', which='both', length=15, width=5)

ax1.spines[:].set_linewidth(4)
ax1.grid()
plt.tight_layout()

plt.savefig(f"../../figures/{power_type}Area_T.png",bbox_inches='tight', dpi=200)
plt.savefig(f"../../figures/jpg/{power_type}Area_T.jpg",bbox_inches='tight')


plt.show()
```

## Area Proportion

```python
def csv_reader(power_type='high'):
    path = os.path.abspath(f'../../../0_Moose_Files/{power_type}/exodus/')
    model_path = path+'/FeNi.csv'
    df = pd.read_csv(model_path,  encoding='utf-8')
    df['area_total'] =  df['area_h1']+df['area_h2']+df['area_h3']+df['area_h4']+df['area_h5']
    # df.tail()
    return df
```

```python
hdf = csv_reader(power_type = 'high_02')
ldf = csv_reader(power_type = 'low_02')
```

```python
hdf.tail()
```

```python
ldf.tail()
```

```python
t_step = len(hdf['time'])

fig, ax1 = plt.subplots(1,1, figsize=(10,8), frameon=False)
ax1.set_facecolor((0, 0, 0, 0))

Low_Ratio = ax1.plot(ldf['time'][:t_step], ldf['area_h4'][:t_step]/ldf['area_h5'][:t_step], color='blue', label=r'1.25 J/mm$^2$', linestyle='dashdot', linewidth=6)
High_Ratio = ax1.plot(hdf['time'][:t_step], hdf['area_h4'][:t_step]/hdf['area_h5'][:t_step], color='red', label=r'5.00 J/mm$^2$', linestyle='dashdot', linewidth=6)

ax1.legend(loc=(0.05, 0.7), ncol=1,  columnspacing=1.0, prop={'weight':'bold', 'size':'25',},  handletextpad=0.5, handlelength=3.0,   frameon=True)#

ax1.set_xlabel(r"t [$\mathbf{\mu s}$] $\mathbf{\longrightarrow}$", weight = 'bold', size=45, color='k')
ax1.set_ylabel(r"Area [$\mathbf{\gamma_{G_2}}$:$\mathbf{\alpha_{G_2}}$]", weight = 'bold', size=45, color='k', labelpad=20)
ax1.get_yaxis().get_offset_text().set_position((-0.01, 0))

ax1.set_yticks([1.8, 2.0, 2.2, 2.4, 2.6, 2.8], labels=[ '1.8', '2.0', '2.2', '2.4', '2.6', '2.8'], size=35, color='k', rotation=0,  weight='bold')
ax1.set_xticks([0, 350, 700, 1050], labels=['0', '350', '750', '1050', ], size=30, color='k', weight='bold')

ax1.set_ylim(1.95, 2.85)
ax1.set_xlim(-50, 1200)
ax1.tick_params(axis='y', direction='inout', which='both', length=15, width=5)
ax1.tick_params(axis='x', direction='inout', which='both', length=15, width=5)

ax1.spines[:].set_linewidth(4)
ax1.grid()
plt.tight_layout()

plt.savefig(f"../../figures/AreaRatio_G2G2.png",bbox_inches='tight', dpi=200)
plt.savefig(f"../../figures/jpg/AreaRatio_G2G2.jpg",bbox_inches='tight')


plt.show()
```
