```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial

# Data
T = np.array([100, 150, 200, 250, 300, 400, 600, 800, 1000, 1200, 1400, 1600, 1800])
rho = np.array([7900, 7890, 7880, 7870, 7860, 7830, 7760, 7690, 7650, 7620, 7520, 7420, 7420])
c = np.array([0.216, 0.324, 0.384, 0.422, 0.450, 0.491, 0.555, 0.692, 1.034, np.nan, np.nan, np.nan, np.nan])*1000
k = np.array([134, 104, 94, 87, 80, 70, 55, 43, 32, 28, 31, np.nan, np.nan])

# Quadratic fits (ignoring nan values)
mask_c = ~np.isnan(c)
mask_k = ~np.isnan(k)

fit_rho = Polynomial.fit(T, rho, 2)
fit_c = Polynomial.fit(T[mask_c], c[mask_c], 2)
fit_k = Polynomial.fit(T[mask_k], k[mask_k], 2)

T_fit = np.linspace(min(T), max(T), 300)
rho_fit = fit_rho(T_fit)
c_fit = fit_c(T_fit)
k_fit = fit_k(T_fit)

# Plot Density
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(T, rho, color="blue", label="Data")
plt.plot(T_fit, rho_fit, "r--", label="Quadratic Fit")
plt.xlabel("Temperature (K)")
plt.ylabel("Density (kg/m³)")
plt.title("Temperature vs Density")
plt.legend()
plt.grid(True)

# Plot Specific Heat
plt.subplot(1, 3, 2)
plt.scatter(T[mask_c], c[mask_c], color="green", label="Data")
plt.plot(T_fit, c_fit, "r--", label="Quadratic Fit")
plt.xlabel("Temperature (K)")
plt.ylabel("Specific Heat (kJ/kg K)")
plt.title("Temperature vs Specific Heat")
plt.legend()
plt.grid(True)

# Plot Thermal Conductivity
plt.subplot(1, 3, 3)
plt.scatter(T[mask_k], k[mask_k], color="purple", label="Data")
plt.plot(T_fit, k_fit, "r--", label="Quadratic Fit")
plt.xlabel("Temperature (K)")
plt.ylabel("Thermal Conductivity (W/m K)")
plt.title("Temperature vs Thermal Conductivity")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

fit_rho.convert().coef, fit_c.convert().coef, fit_k.convert().coef
```
