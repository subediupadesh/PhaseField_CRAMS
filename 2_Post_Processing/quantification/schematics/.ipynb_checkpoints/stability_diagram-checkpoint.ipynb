```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import matplotlib.patches as patches
from matplotlib.patches import Polygon 
from scipy.io import loadmat
from matplotlib.colors import ListedColormap
from matplotlib.lines import Line2D
from matplotlib.text import Text
```

```python
def Gibbs_LIQ(c_temp):
    c, T = c_temp
    F, A, c0, B, T0, D = 6983.4, 8545, 0.94, 0.00003, 1975, 20
    return F*(A*(c-c0)**2 + B*(T-T0)**2  - D)

def Gibbs_FCC(c_temp):
    c, T = c_temp
    F, A, c0, B, T0, D = 7932.8, 7125, 0.88, 0.00003, 1280.5, 15
    return F*(A*(c-c0)**2 + B*(T-T0)**2  - D)

def Gibbs_BCC(c_temp):
    c, T = c_temp
    F, A, c0, B, T0, D = 2397.7, 7560, 0.96, 0.00003, 881, 10
    return F*(A*(c - c0)**2 + B*(T - T0)**2 - D)
```

## High Power

```python
power_level = 'high' #'low'

numpy_path = os.path.abspath(f'../../../1_Simulation_Results/numpy_files/{power_level}_power/')

colormap_data = loadmat('../cold2hot.mat')
cold2hot = ListedColormap(colormap_data['colormap'])
hot2cold = ListedColormap(colormap_data['colormap'][::-1])
```

```python
comp = np.load(numpy_path+'/comp.npy')
# LIQ1 = np.load(numpy_path+'/LIQ1.npy')
# FCC2 = np.load(numpy_path+'/FCC2.npy')
# BCC3 = np.load(numpy_path+'/BCC3.npy')
# FCC4 = np.load(numpy_path+'/FCC4.npy')
# BCC5 = np.load(numpy_path+'/BCC5.npy')
temp = np.load(numpy_path+'/temp.npy')
time = np.load(numpy_path+'/time.npy')
time = time*1e-6 # mu_sec
Nicomp = 1 - comp
```

```python
scan_speed = r'300 $\mathbf{mm/s}$'
power = r"45 $\mathbf{W}$"
t_step = 276  ## Actual time for consistency to be analysed at t = 550 mu_s & laser at 240.5 mu_m 
    
laser_speed = 300e-3 # m/s
scaling_factor = 1
if time[t_step] <= 15e-6:
    laser_act_pos = 80e-6 * 1e6 # convering to meters
else:
    laser_act_pos = (80e-6 + (time[t_step]-15e-6)*laser_speed) * 1e6  ## converting to visualization unit 
laser_pos = laser_act_pos*scaling_factor # Laser position in accordance with figure

fig, ax1 = plt.subplots(1,1, figsize=(16,8), frameon=False)
ax1.set_facecolor((0, 0, 0, 0))

# gist_rainbow
cmap =  plt.get_cmap('hsv')
cmap.set_under('white', alpha=0)
hmap1 = ax1.imshow(Nicomp[:,:,t_step], extent=[0, 500, 0, 90], cmap=cmap, vmax=1-0.875, vmin=1-0.966, aspect=1, origin="upper", interpolation='bilinear')

ax1.arrow(laser_pos, -51, 0, 50,  width = 4.5, color='red', length_includes_head=True, clip_on=False)
ax1.text(laser_pos-20,-12, power, rotation=90, color='k', fontsize=20, weight='bold')

if laser_pos < 250:
    ax1.set_title(f't = {time[t_step]*1e6:.0f} '+r'$\mathbf{\mu}$s', pad=10, loc='right', fontsize=25, weight='bold')
else:
    ax1.set_title(f't = {time[t_step]*1e6:.0f} '+r'$\mathbf{\mu}$s', pad=10, loc='left', fontsize=25, weight='bold')        

ax1.text(laser_pos+10, -40, s=f'{scan_speed}', fontsize=20, weight = 'bold')
ax1.arrow(laser_pos+10, -28, 30, 0,  width = 3, color='k', fill=False, clip_on=False)

ax1.set_xlim(-1, 500)
ax1.set_ylim(91, -1)

ax1.set_xlabel('$\mathbf{\longleftarrow}$                 500 $ \mathbf{\mu m}$                $\mathbf{\longrightarrow}$', fontsize = 27, weight ='bold',)
ax1.annotate(r'$\mathbf{\leftarrow}$90 $\mathbf{\mu m}$ $\mathbf{\rightarrow}$', xy=(0.5, 0.5), xytext=(-0.025, 0.5), rotation=90,
             xycoords='axes fraction', textcoords='axes fraction', fontsize = 27, weight = 'bold', color='k', ha='center', va='center')

ax1.tick_params(axis='both', labelcolor='black', labelsize=65, bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)
ax1.spines[:].set_linewidth(2)

# LIQ1_txt = ax1.text(217, 19, r'$\mathbf{LIQ}$', fontsize = 35, weight='bold', color='white')
# FCC2_txt = ax1.text(450, 20, r'$\mathbf{\gamma_{G_1}}$', fontsize = 40, weight='bold', color='lime')
# FCC4_txt = ax1.text(320, 60, r'$\mathbf{\gamma_{G_2}}$', fontsize = 40, weight='bold', color='lime')
# BCC3_txt = ax1.text(50, 30, r'$\mathbf{\alpha_{G_1}}$', fontsize = 40, weight='bold', color='k')
# BCC5_txt = ax1.text(125, 87, r'$\mathbf{\alpha_{G_2}}$', fontsize = 30, weight='bold', color='k')

contour_levels = [677, 1468,  1864, ]
label_colors = ['k', 'k', 'k', 'white']
contour = ax1.contour(temp[:, :, t_step], extent=[0, 500, 0, 90], levels=contour_levels, linewidths=5, origin="upper", colors=['lime', 'k', 'red'], linestyles='dashdot')
[(label.set_fontweight('bold'), label.set_color(color)) for label, color in zip(plt.clabel(contour, inline=True, fontsize=18, fmt='%1.0f K'), label_colors)]


############## Color Bar ###########
ax2 = fig.add_axes([0.94, 0.355, 0.05, 0.278]) 
cbar = fig.colorbar(hmap1, cax=ax2, orientation='vertical')
cbar.ax.tick_params(labelsize=25, direction='in', length=0) 
text = Text(x=0.907, y=0.38, text=r'$\mathbf{\leftarrow}$ c$_\mathbf{Ni}$ $\mathbf{\rightarrow}$', fontsize=30, color='k', rotation=90, weight='bold')
fig.add_artist(text)
# cbar.set_ticks([300, (np.max(temp[:,:,t_step])+300)/2, np.max(temp[:,:,t_step])], labels=[300, f'{(np.max(temp[:,:,t_step])+300)/2 :.0f}', f'{np.max(temp[:,:,t_step]):.0f}'], weight='bold') 
cbar.set_ticks([0.04, 0.08, 0.12], labels=[0.04, 0.08, 0.12], weight='bold') 

ax1.tick_params(axis='both', labelcolor='black', labelsize=65, bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)
ax1.spines[:].set_linewidth(2)
ax1.spines[:].set_color('k')
ax2.spines[:].set_linewidth(2)
line = Line2D([1.05, 1.05], [0.35, 0.65], transform=fig.transFigure, color="none", linewidth=3, linestyle='-')
fig.add_artist(line)


# plt.savefig(f"../../figures/{power_level}Composition_{t_step}.png",bbox_inches='tight', dpi=200)
# plt.savefig(f"../../figures/jpg/{power_level}Composition_{t_step}.jpg",bbox_inches='tight')

plt.show()
```
